```python
import pandas as pd    #A
import matplotlib.pyplot as plt    #B

# Example dataset: Phishing attempts recorded over 12 months
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],    #C
        'Phishing_Attempts': [23, 25, 30, 35, 40, 45, 50, 60, 75, 80, 85, 90]}    #C

df = pd.DataFrame(data)    #D

# Plotting the trend
plt.figure(figsize=(10, 6))    #E
plt.plot(df['Month'], df['Phishing_Attempts'], marker='o', linestyle='-', color='blue')    #F
plt.title('Monthly Trend of Phishing Attempts')    #G
plt.xlabel('Month')    #H
plt.ylabel('Number of Phishing Attempts')    #I
plt.grid(True)    #J
plt.show()    #K
```
